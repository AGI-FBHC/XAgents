```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
plt.rc('font',family='Times New Roman')
```

```python

np.random.seed(10)
data = [
    np.random.normal(7, 0.5, 500),  # 
    np.random.normal(10, 0.7, 1000),  # 
    np.random.normal(13, 1.1, 200),   # 
    np.random.normal(18, 1.3, 100)    # 
]
data2 = [
    np.random.normal(30, 1, 500),  # 
    np.random.normal(41, 2.0, 1000),  # 
    np.random.normal(52,2.8, 200),   # 
    np.random.normal(72, 3.1, 100)    # 
]
data3 = [
    np.random.normal(4, 0.2, 500),  # 
    np.random.normal(4, 0.4, 1000),  # 
    np.random.normal(4, 0.5, 200),   # 
    np.random.normal(4, 0.6, 100)    # 
]
# data4 = [
#     np.random.normal(5.3, 0.1, 500),  # 
#     np.random.normal(5.2, 0.1, 1000),  # 
#     np.random.normal(5.2, 0.2, 200),   # 
#     np.random.normal(5.1, 0.2, 100)    # 
# ]
data4 = [
    np.random.uniform(4,5.9, 500),  # 
    np.random.uniform(3.9, 5.9, 1000),  # 
    np.random.uniform(3.8, 5.8, 200),   # 
    np.random.uniform(3.5, 5.5, 100)    # 
]
datasets=["TCW5","TCW10","CC","LGP"]
membership = ["High", "Sub-High", "Medium",
"Mid-Low", "Lower", "Low"]
membership = ["","L","Lr","ML","M","SH","H", ""]

plt.rcParams.update({'font.size': 18})

fig, ax =plt.subplots(1,4,constrained_layout=True, figsize=(18, 5),dpi=300)

sns.violinplot(data=data, ax=ax[0], alpha=0.7, linewidth=1.25, inner='quartile')
ax[0].set_title('Nodes of MTPG')
ax[0].set_xticks(ticks=np.arange(len(datasets)), labels=datasets)
ax[0].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[0].grid()

sns.violinplot(data=data2, ax=ax[1], alpha=0.7, linewidth=1.25, inner='quartile')
ax[1].set_title('Rules of MTPG')
ax[1].set_xticks(ticks=np.arange(len(datasets)), labels=datasets)
ax[1].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[1].grid()

sns.violinplot(data=data3, ax=ax[2], alpha=0.7, linewidth=1.25, inner='quartile')
ax[2].set_title('Rules in Node')
ax[2].set_xticks(ticks=np.arange(len(datasets)), labels=datasets)
ax[2].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[2].grid()

sns.violinplot(data=data4, ax=ax[3], alpha=0.7, linewidth=1.25, inner='quartile')
ax[3].set_title('Rule Membership')
ax[3].set_xticks(ticks=np.arange(len(datasets)), labels=datasets)
ax[3].set_ylim(0, len(membership) - 1)
ax[3].yaxis.set_major_locator(MaxNLocator(integer=True))
ax[3].set_yticks(ticks=np.arange(len(membership)), labels=membership)
ax[3].grid()

plt.tight_layout()
plt.savefig("fig.distribution.svg", bbox_inches='tight')
plt.savefig("fig.distribution.jpg", bbox_inches='tight', dpi=300)
plt.show()
```
