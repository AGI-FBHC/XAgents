```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
plt.rc('font',family='Times New Roman')
```

```python
# TCW5 TCW10 CC LGP 
# scores=[[84.4,71.5,57.7],[88.1,72.5,61.7],[75,67.5,57.7],[83.5,72.5,57.7]]
scores=[[84.4,88.1,83.5,75],[71.5,72.5,72.5,67.5],[57.7,61.7,57.7,57.7]]
errs=[[2.7,3.7,1.7,2.7],[1.7,4.7,4.7,1.7],[5.7,5.7,5.7,5.7]]
# errs=[[2.7,1.7,5.7],[3.7,4.7,5.7],[1.7,4.7,5.7],[2.7,1.7,5.7]]
label=['XAgents',"-ITRDM","-MTPG"]
x_tick=["TCW5","TCW10","CC","LGP"]
color=['royalblue','skyblue','orange']
error_attri = dict(elinewidth = 2, ecolor="black", capsize = 3)

plt.rcParams.update({'font.size': 24})
plt.subplots(1,1,figsize=(10,6),dpi=300)
error_attri = dict(elinewidth = 1, ecolor="black", capsize = 3)

x=range(4)
bar_width = 0.3

plt.subplot(1,1,1)
plt.bar(x, scores[0],width=bar_width, yerr=errs[0],  error_kw=error_attri,color=color[0],alpha=0.7, label=label[0])
plt.bar([i + bar_width for i in x], scores[1],width=bar_width, yerr=errs[1],  error_kw=error_attri,color=color[1],alpha=0.7, label=label[1])
plt.bar([i + bar_width*2 for i in x], scores[2],width=bar_width, yerr=errs[2],  error_kw=error_attri,color=color[2],alpha=0.7, label=label[2])
plt.xticks([i + bar_width for i in x], x_tick)
plt.ylim(50, 100)
plt.ylabel("Score")
plt.grid()
plt.legend(ncol=3)

plt.tight_layout()
plt.savefig("fig.ablation.svg", bbox_inches='tight')
plt.savefig("fig.ablation.jpg", bbox_inches='tight', dpi=300)
plt.show()
```
